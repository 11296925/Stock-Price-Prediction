```python
import glob
import re
import sys

from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

sentim_analyzer = SentimentAnalyzer()
pos, neg = 0, 0
allpos = 0

files = glob.glob("NewsArticles/AAPL/*.txt")
for file in files:
    wordList, relevant_list = [], []
    f = open(file, "r")
    for i in f:
        if "AAPL" in i:
            relevant_list.append(i)
    for text in relevant_list:
        if len(text) > 100:
            print(text)
            words = re.sub("[^\w]", " ",  text).split()
            words = util.mark_negation(words)
            for word in words:
                pos += sid.polarity_scores(word)['pos']
                print(pos)
                neg += sid.polarity_scores(word)['neg']
                print(neg)
print(pos/(pos+neg))

# print(pos)
# print(neg)
#     if totwords>100:
#         print(file)
#         print(pos-neg)
# 

# print(allpos)
            
#     obj = Sentiment(text=wordList)
#     data = obj.get
#     print(data)
```

```python
pos, neg = 0, 0

files = glob.glob("NewsArticles/GoogleNews/*.txt")
for file in files:
    wordList = []
    f = open(file, "r")
    lines = f.readlines()
    for line in lines:
        words = re.sub("[^\w]", " ",  line).split()
        words = util.mark_negation(words)
        for word in words:
            pos += sid.polarity_scores(word)['pos']   
            neg += sid.polarity_scores(word)['neg']
print('Positive rate:', pos)
print('Negative rate:', neg)
print('Ratio positive/negative = ', pos/neg)
```
